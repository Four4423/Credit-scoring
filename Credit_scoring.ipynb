```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import pyarrow.parquet as pq


df = pd.DataFrame()

for i in range(0, 2):
    file_path = f"*file_path*/train_data/train_data_{i}.pq"
    parquet_file = pq.ParquetFile(file_path)
    for j in range(parquet_file.num_row_groups):
        df_temp = parquet_file.read_row_group(j).to_pandas()
        df = df._append(df_temp, ignore_index=True)
df
```

```python
target = pd.read_csv("C:/Users/Konstantin/Downloads/alfaTest/train_target.csv")
```

```python
merged_data = pd.merge(df, target, on='id')
merged_data
```

```python
X = merged_data.drop(['flag'], axis=1)
y = merged_data['flag']
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X, y)
```

```python
test1 = pd.read_csv("*file_path*")
test2 = pd.read_csv("*file_path*")

# ÐžÐ±ÑŠÐµÐ´Ð¸Ð½ÐµÐ½Ð¸Ðµ Ð´Ð°Ð½Ð½Ñ‹Ñ… Ð¿Ð¾ id
merged_data_test = pd.merge(test1, test2, how = 'outer')
merged_data_test
```

```python
predictions = model.predict_proba(merged_data_test)
predictions
```

```python
predictions_first = [p[1] for p in predictions]
```

```python
merged_data_test['prediction'] = predictions_first
merged_data_test
```

```python
df_grouped = merged_data_test.groupby('id').agg({'prediction': 'mean'}).reset_index()

print(df_grouped)
```
